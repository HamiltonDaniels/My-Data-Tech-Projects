# 5. DATA VISUALISATION USING MATPLOTLIB
**Step 6. Plot a line graph**

```python
df_cr7['Date'] = pd.to_datetime(df_cr7['Date'],errors='coerce')
df_cr7 = df_cr7.sort_values(by='Date')
plt.figure(figsize=(10, 6))
plt.plot(df_cr7['Date'], df_cr7['At_score'], label='At Score', marker='o')
plt.xlabel('Date')
plt.ylabel('score at time')
plt.title('Score Over Time (Time-series)')
plt.xticks(rotation=45)
plt.legend

plt.show
```

```python
import plotly.express as px

px.histogram(
    df_cr7,
    x='Club',
    title="Goals per Clubs - Seasons",
    log_x=False,
    log_y=False,
    #symbol='title',
    #markers=True,
    #width=800, 
    height=500,
    color='Season',
    hover_name='Season',
    hover_data=['Competition','Season','Club'])
```

```python
assists = df_cr7.groupby('Goal_assist').size().reset_index(name = 'assists').sort_values(by = 'assists', ascending= False)[:5]
fig = px.bar(assists, x='Goal_assist', y='assists', title="CR7's top assists",
            labels={'Goal_assist': 'Player', 'assists': 'Number of Assists'}, text= 'assists',
            color = 'Goal_assist')

fig.show()
```

```python
px.histogram(
    df_cr7,
    x='Competition',
    title="Goals per competition",
    log_x=False,
    log_y=False,
    #symbol='title',
    #markers=True,
    #width=800, 
    height=500,
    color='Club',
    hover_name='Club',
    hover_data=['Competition','Club'])
```

**Step 10. Create a rolling average for time-series analysis**

```python
df_cr7['minute_avg'] = df_cr7['Season'].rolling(window=5).mean()
plt.plot(df_cr7['date'], df_cr7['minute_avg'], label='Minute Average')
plt.legend()
plt.show()
```
